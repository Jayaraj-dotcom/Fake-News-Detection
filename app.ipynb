```python
import pandas as pd
from sklearn.model_selection import train_test_split 
from sklearn.feature_extraction.text import TfidfVectorizer 
from sklearn.linear_model import LogisticRegression 
from sklearn.metrics import classification_report 
import re 
import joblib 
import string
```

```python
fake = pd.read_csv('Fake.csv') 
true = pd.read_csv('True.csv')
```

```python
fake['class'] = 0
true['class'] = 1
```

```python
data = pd.concat([fake,true], axis = 0)
```

```python
data.sample(10)
```

```python
data = data.drop(["title", "subject", "date"], axis = 1) 
```

```python
data.reset_index(inplace=True)
```

```python
data.drop(['index'], axis = 1,inplace=True)
```

```python
data.sample(5)
```

```python
# counting the number of missing values in the dataset
data.isnull().sum()
```

```python
# replacing the null values with empty string
data = data.fillna('')
```

```python
def clean_text(text):
    text = text.lower()
    text = re.sub("\[.*?\]", "", text)
    text = re.sub("\\W"," ",text)
    text = re.sub("https?:://\S+/www\.\S+","", text)
    text = re.sub("<.*?>+","",text)
    text = re.sub("[%s]" % re.escape(string.punctuation),"",text)
    text = re.sub("\n","",text)
    text = re.sub("\w*\d\w*","", text)
    return text 
```

```python
data["text"] = data["text"].apply(clean_text)
```

```python
x=data["text"] 
y=data["class"]

xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=0.25, random_state=42)
```

```python
vectorizer = TfidfVectorizer() 
xv_train = vectorizer.fit_transform(xtrain) 
xv_test = vectorizer.transform(xtest)
```

```python
lr = LogisticRegression() 
lr.fit(xv_train, ytrain)
```

```python
prediction = lr.predict(xv_test) 
lr.score(xv_test,ytest)
```

```python
print(classification_report (ytest, prediction))
```

```python
joblib.dump(vectorizer, "vectorizer.jb") 
joblib.dump(lr,"lr_model.jb")
```

```python
# Check the size of training and testing sets
train_size = len(xtrain)
test_size = len(xtest)
total_size = train_size + test_size

# Calculate percentage split
train_percentage = (train_size / total_size) * 100
test_percentage = (test_size / total_size) * 100

print(f"Training Set Size: {train_size} ({train_percentage:.2f}%)")
print(f"Testing Set Size: {test_size} ({test_percentage:.2f}%)")
```
